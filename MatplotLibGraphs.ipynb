```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Day')          # x-axis label
plt.ylabel('Temperature')  # y-axis label
plt.title("Weather")       # Title for graph
x=np.array([1,2,3,4,5,6,7])
y=np.array([50,51,52,48,47,49,46])
plt.plot(x,y)
```

```python
plt.plot(x,y,color='green',alpha=0.5) # alpha defines the transparency of the line (the value of alpha lies between 0 and 1)
```

```python
################# ANOTHER DATASET ###################
days=np.array([1,2,3,4,5,6,7])
max_t=np.array([50,51,52,48,47,49,46])
min_t=np.array([43,42,40,44,33,35,37])
avg_t=np.array([45,48,48,46,40,42,41])
```

```python
############ ADDING LEGENDS TO GRAPH #################
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='Max Temperature')           # label is given so that each lagend is labeled
plt.plot(days,min_t, label='Min Temperature')
plt.plot(days,avg_t, label='Average Temperature')

plt.legend()                               # This adds the legends to the graph
# plt.legend(loc="upper right")           To change the location of legend, by default the property is set to best

plt.grid()                     # Used to exactly figure out where the point lies

###########  Plotting 3 graphs in one only #################
```

```python
################# BAR CHAT #########################
company=['GOOGLE','AMAZON','MICROSOFT','FB']
revenue=np.array([90,136,89,27])
profit = np.array([40,2,34,12])
```

```python
# to plot a bar chat we use plt.bar(arg1,arg2), both the args should be in int and not in strings so to convert company
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)  # this functions replaces the values with corresponding elements from your list
plt.ylabel("revenue(bln)")
plt.title("US Tech Stcoks")
plt.bar(ypos-0.2,revenue,width=0.4, label='Revenue')
plt.bar(ypos+0.2,profit,width=0.4, label='Profit')

plt.legend()
```

```python
############### Horizontal Bar Chats #####################
plt.yticks(ypos,company)
plt.title("US TEch Stcoks")
plt.barh(ypos-0.2,revenue, label='Revenue')
plt.legend()
```

```python
############################# Histograms ######################################
blood_sugar = np.array([113,85,90,150,149,88,93,115,135,80,77,82,129])
plt.hist(blood_sugar,bins=[80,100,125,150], rwidth=0.95)  # By default the value of bins is 10
```

```python
blood_sugar_men = np.array([113,85,90,150,149,88,93,115,135,80,77,82,129])
blood_sugar_women = np.array([67,98,89,120,133,150,84,69,89,79,120,112,100])
plt.xlabel('suagr range')
plt.ylabel('Total No of Patients')
plt.title('Blood Suagr Analysis')
plt.hist([blood_sugar,blood_sugar_women],bins=[80,100,125,150], rwidth=0.95,color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
###################### PIE CHART ################################
exp_vals=np.array([1400,600,300,410,250])
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
plt.axis("equal")      # for perfect circle
plt.pie(exp_vals,labels=exp_labels,radius=1, autopct='%0.2f%%', explode=[0,0,0.5,0,0]) #autopct shows the percentage of each pie. Explode will show that specific pie differently
plt.show()             # to show only pie chart
```
