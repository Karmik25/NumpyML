```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Tutorial 5.csv',parse_dates=["day"])
df.set_index('day',inplace=True)
df
```

```python
# FILL NA
new_df = df.fillna(0)      # replaces the NaN values with 0
new_df
```

```python
# if we want to fill for NaN for some fields only, we enter a dictionary in fillna
new_df = df.fillna({
    'temperature': 0,
    'windspeed' : 0,
    'event' : 'no event'
})
new_df
```

```python
#2
new_df = df.fillna(method="ffill", limit=1) #this method carry forwards the previous values i.e ffill means forward fill (we can also use bfill = backward fill)
                                            #limit - limits the number of time the data should be copied
new_df
```

```python
#3
new_df = df.fillna(method='ffill',axis='columns')  # copies the value of columns (axis)
new_df
```

```python
#4 Interpolate
new_df = df.interpolate()   # comes with the best guess for the NaN values
new_df
```

```python
#5
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how="all")  # this preseveres the columns with some data, unlike the previous one which removes all the columns with NaN
new_df
```

```python
new_df = df.dropna(thresh=2)  #rows with atleast 2 valid data will be displayed
new_df
```
