```python
import pandas as pd
import numpy as np
```

```python
india_weather = pd.DataFrame({
    "city" : ["Mumbai","Delhi","Bangalore"],
    "temperatur" : [32,45,30],
    "humidity" : [80,60,78]
})
india_weather
```

```python
usa_weather = pd.DataFrame({
    "city" : ["New York","Chicago","Orlando"],
    "temperatur" : [21,14,35],
    "humidity" : [68,65,75]
})
usa_weather
```

```python
# concating the two dataframes
df = pd.concat([india_weather,usa_weather], ignore_index = True)
df
```

```python
#to differentiate between the two dataframes
df = pd.concat([india_weather,usa_weather], keys=["india","USA"])
df
```

```python
# appending the second dataframe as column
temperature_df = pd.DataFrame({
    "city" : ["Mumbai","Delhi","Bangalore"],
    "temperature" : [32,45,30],
},index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city" : ["Delhi","Mumbai"],
    "windspeed" : [7,12],
},index=[1,0])
windspeed_df
```

```python
new_df = pd.concat([temperature_df,windspeed_df],axis = 1)
new_df
```
