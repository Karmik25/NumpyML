```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Tutorial 4.csv")
df
```

```python
#1
new_df = df.replace(-99999,np.NaN)  #the replace function takes 2 parameters(first is the values to be replaced and second is the value we want to replace)
                                    #we can also enter a list as first paramter
                                    # dictionary can also we used as first parameter
new_df
```

```python
new_df = df.replace({
    'temperature': -99999,
    'windspeed' : -99999,
    'event' : '0'
},np.NaN)
new_df
```
